## **Extracting Exports Data - Census.gov API**

```python
import requests
import pandas as pd


# Census.gov API call
base_url = "https://api.census.gov/data/timeseries/intltrade/exports/hs"

years = range(2013, 2025)
exports = []

for year in years:
    params = {
        "get": "CTY_NAME,CTY_CODE,ALL_VAL_YR",
        "time": f"{year}-12"
    }

    response = requests.get(base_url, params=params)

    data = response.json()
    df = pd.DataFrame(data[1:], columns=data[0])
    df["year"] = year
    exports.append(df)


```

```python
# Overview of Extractions
exports
```

```python
# Merging Data over Years
exports_df = pd.concat(exports, ignore_index=True)

# Converting values to Numeric
exports_df["ALL_VAL_YR"] = pd.to_numeric(exports_df["ALL_VAL_YR"])

# Getting Data Overview
print(exports_df.head())
```

```python
exports_df.to_csv("exports_df.csv")
```

## **Extracting Imports Data - Census.gov API**

## ## Monthly Data Collection — Inflation & Interest Rates

```python
# Importing libraries
import requests
import pandas as pd
import time
```

```python
# Calling IMF API for monthly inflation data (Consumer Price Index, percent change)
url = "https://sdmxapi.data.imf.org/v2/data/dataflow/IMF.STA/IFS/1.0/ALL.PCPI_PC_PP_PT.M"

params = {
    "startPeriod": "2013-01",   # from January 2013
    "endPeriod": "2024-12",     # to December 2024
    "format": "csv"             # we want CSV format back
}

response = requests.get(url, params=params, timeout=120)

# Checking
print("Status code:", response.status_code)
print("Response size:", len(response.text), "characters")
```
